```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def cov2d_single_channel(img, kernal, stride=1, padding=0):
    H, W = img.shape
    k = kernal.shape[0]
    imgp = np.pad(img, padding, mode='constant')
    H0 = (H + 2*padding - k) // stride + 1
    W0 = (W + 2*padding - k) // stride + 1
    out = np.zeros((H0, W0))
    for i in range(H0):
        for j in range(W0):
            patch  = imgp[i*stride : i*stride+k, j*stride : j*stride +k]
            out[i,j] = np.sum(patch*kernal)
    return out
```

```python
H,W = 20, 100
img = np.hstack(
    np.ones((H, W//4))*50,
    np.ones((H, W//4))*200,
    np.ones((H, W//4))*50,
    np.ones((H, W//4))*200
)
# verirocal edge-detection kernal (sobel-like)
kernal = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])
out = cov2d_single_channel(img, kernal, stride=1, padding=0)

out_display = (out - out.min()) / (out.max() - out.min())

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title('input image with vertical Edge')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(out_display, cmap='gray')
plt.title('After Convoloution (Edge Map)')
plt.axis('off')
plt.show()
```
